```python
import pandas as pd
```

```python
dados = pd.read_csv("State_of_data_2022.csv", sep=",")
```

```python
dados.head()
```

```python
lista = dados.columns.tolist()

divisao = {
    'P0': ('Parte0', 'Identificador'),
    'P1': ('Parte1', 'Dados demográficos'),
    'P2': ('Parte2', 'Dados sobre carreira'),
    'P3': ('Parte3', 'Desafios dos gestores de times de dados'),
    'P4': ('Parte4', 'Conhecimentos na área de dados'),
    'P5': ('Parte5', 'Objetivos na área de dados'),
    'P6': ('Parte6', 'Conhecimentos em Engenharia de Dados/DE'),
    'P7': ('Parte7', 'Conhecimentos em Análise de Dados/DA'),
    'P8': ('Parte8', 'Conhecimentos em Ciências de Dados/DS')
}

# Extrair os dados da lista
dados_extratos = [(item.split("'")[1], item.split("'")[3]) for item in lista]

# Criar DataFrame
descricao = pd.DataFrame(dados_extratos, columns=['IdentificadorQuestao', 'Questao'])

# Adicionar coluna de divisão do questionário
descricao['DivisaoQuestionario'] = descricao['IdentificadorQuestao'].apply(lambda x: divisao.get(x.split('_')[0], ('Outro', 'Outro'))[0])

# Adicionar coluna de identificador
descricao['NomeCategoria'] = descricao['IdentificadorQuestao'].apply(lambda x: divisao.get(x.split('_')[0], ('Outro', 'Outro'))[1])

# Reorganizar colunas
descricao = descricao[['DivisaoQuestionario', 'NomeCategoria', 'IdentificadorQuestao', 'Questao']]

# Erros encontrados
descricao.loc[descricao['IdentificadorQuestao'] ==  'P7_1 ', 'IdentificadorQuestao'] = "P7_a"
descricao.loc[descricao['IdentificadorQuestao'] ==  'P8_3 ', 'IdentificadorQuestao'] = "P8_c"

# Exibir o DataFrame
descricao.head()
```

# Fazendo coluna pergunta

```python
def extrair_pergunta(identificador):
    partes = identificador.split('_')
    if len(partes) > 2:
        return '_'.join(partes[:2]).strip()
    else:
        return identificador.strip()

# Aplicando a função à coluna "IdentificadorQuestao" para criar a nova coluna "pergunta"
descricao['pergunta'] = descricao['IdentificadorQuestao'].apply(extrair_pergunta)
```

## Fazendo nova coluna

```python
descricao_filtrada = descricao[descricao['IdentificadorQuestao'].str.len() <= 5]
descricao_filtrada
```

```python
dicionario = descricao_filtrada.set_index('pergunta')['Questao'].to_dict()
```

```python
dicionario
```

```python
# Função para mapear as perguntas para as descrições correspondentes
def mapear_descricao(pergunta):
    return dicionario.get(pergunta.strip(), 'Sem descrição disponível')

# Criando a coluna "DescricaoPergunta" usando .loc para evitar SettingWithCopyWarning
descricao.loc[:, 'DescricaoPergunta'] = descricao['pergunta'].map(mapear_descricao)
descricao
```

## Apagando repetições entre Questão e DescricaoPergunta

```python
repeticoes = descricao[descricao['Questao'] == descricao['DescricaoPergunta']]
repeticoes
```
