```python
import pandas as pd
```

# Importando CSV

```python
dados = pd.read_csv("State_of_data_2022.csv", sep=",")
```

```python
dados.head()
```

# Extraindo os dados da lista e criando DataFrame

```python
lista = dados.columns.tolist()

divisao = {
    'P0': ('Parte0', 'Identificador'),
    'P1': ('Parte1', 'Dados demográficos'),
    'P2': ('Parte2', 'Dados sobre carreira'),
    'P3': ('Parte3', 'Desafios dos gestores de times de dados'),
    'P4': ('Parte4', 'Conhecimentos na área de dados'),
    'P5': ('Parte5', 'Objetivos na área de dados'),
    'P6': ('Parte6', 'Conhecimentos em Engenharia de Dados/DE'),
    'P7': ('Parte7', 'Conhecimentos em Análise de Dados/DA'),
    'P8': ('Parte8', 'Conhecimentos em Ciências de Dados/DS')
}

dados_extratos = [(item.split("'")[1], item.split("'")[3]) for item in lista] # Extraindo

descricao = pd.DataFrame(dados_extratos, columns=['IdentificadorQuestao', 'Questao']) # Criando DataFrame
#Questão será uma coluna temporária

#removendo espaços que ficaram no final das strings da coluna 'IdentificadorQuestao'
descricao['IdentificadorQuestao'] = descricao['IdentificadorQuestao'].str.rstrip()

descricao.head() 
```

# Adicionando coluna sonbre a divisão do questionário

```python
descricao['DivisaoQuestionario'] = descricao['IdentificadorQuestao'].apply(lambda x: divisao.get(x.split('_')[0],
                                                                                                 ('Outro', 'Outro'))[0])

descricao.head()
```

# Adicionando coluna com o nome da categoria

```python
descricao['NomeCategoria'] = descricao['IdentificadorQuestao'].apply(lambda x: divisao.get(x.split('_')[0],
                                                                                           ('Outro', 'Outro'))[1])

descricao.head()
```

#  Erros encontrados no CSV durante o processo

## Tabela com esses erros

```python
identificadores_selecionados = ['P7_1', 'P8_3', 'P4_i']
erros = descricao.loc[descricao['IdentificadorQuestao'].isin(identificadores_selecionados)]

erros
```

## Arrumando os erros

```python
# Identificador escrito errado
descricao.loc[descricao['IdentificadorQuestao'] ==  'P7_1', 'IdentificadorQuestao'] = "P7_a" 
descricao.loc[descricao['IdentificadorQuestao'] ==  'P8_3', 'IdentificadorQuestao'] = "P8_c"

#Pergunta esta faltando
descricao.loc[descricao['IdentificadorQuestao'] ==  'P4_i', 'Questao'] = "Dentre as opções listadas, qual sua ferramenta de visualização preferida?"
```

# Aplicando a função à coluna "IdentificadorQuestao" para criar a nova coluna "pergunta"

```python
def extrair_pergunta(identificador):
    partes = identificador.split('_')
    if len(partes) > 2:
        return '_'.join(partes[:2]).strip()
    else:
        return identificador.strip()

descricao['pergunta'] = descricao['IdentificadorQuestao'].apply(extrair_pergunta)

descricao.head()
```

# Criando dicionario e tabela descricao_filtrada para auxiliar na criação da coluna 'DescricaoPergunta'

```python
descricao_filtrada = descricao[descricao['IdentificadorQuestao'].str.len() <= 5] #somente identificadores com 5 caracteres

descricao_filtrada.head()
```

```python
dicionario = descricao_filtrada.set_index('pergunta')['Questao'].to_dict()

dicionario
```

# Usando função para mapear as perguntas para as descrições correspondentes

```python
def mapear_descricao(pergunta):
    return dicionario.get(pergunta.strip(), 'Sem descrição disponível')

# Criando a coluna "DescricaoPergunta" usando .loc para evitar SettingWithCopyWarning
descricao.loc[:, 'DescricaoPergunta'] = descricao['pergunta'].map(mapear_descricao)

descricao.head()
```

# Criando coluna para exibir o número da identificador das opções para alguma pergunta quando existir.

```python
def extrair_opcao(questao):
    partes = questao.split('_')
    if len(partes) >= 3:
        return partes[2]
    else:
        return '-'


descricao.loc[:, 'Opcao'] = descricao['IdentificadorQuestao'].apply(lambda x: extrair_opcao(x))

descricao.head()
```

# Função para definir a descrição da opção

```python
def descricao_opcao(opcao, questao):
    if opcao == '-':
        return '-'
    else:
        return questao
    
descricao.loc[:, 'DescricaoOpcao'] = descricao.apply(lambda row: descricao_opcao(row['Opcao'], row['Questao']), axis=1)

descricao.head()
```

# Apagando coluna temporária 'Questao'

```python
descricao = descricao.drop(columns=['Questao'])

descricao.head()
```

# Definindo coluna sobre o tipo da pergunta. Se é fechada, aberta ou uma linha de opção de uma pergunta aberta

```python
#Marcando primeirando todas como aberta
descricao.loc[:, 'TipoPergunta'] = 'Aberta'

# Identificando as perguntas fechadas
perguntas_fechadas = descricao['pergunta'].value_counts() == 1
perguntas_fechadas = perguntas_fechadas[perguntas_fechadas].index.tolist()

# Marcando as perguntas fechadas como 'Fechada'
descricao.loc[descricao['pergunta'].isin(perguntas_fechadas), 'TipoPergunta'] = 'Fechada'

# Marcando as perguntas repetidas como 'Opção da Pergunta [dado da coluna 'Pergunta']'
descricao.loc[descricao['pergunta'].duplicated(keep='first'), 'TipoPergunta'] = 'Opção da Pergunta ' + descricao['pergunta']

# Exibir DataFrame resultante
descricao.head()
```

# Reorganizando colunas

```python
nova_ordem_colunas = ['IdentificadorQuestao', 'DivisaoQuestionario', 'NomeCategoria', 'pergunta','TipoPergunta', 'DescricaoPergunta', 'Opcao', 'DescricaoOpcao']

# Reorganizar as colunas
descricao = descricao[nova_ordem_colunas]

descricao.head()
```
